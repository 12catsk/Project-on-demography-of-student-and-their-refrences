```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = './cleaned_data.csv'
df = pd.read_csv(file_path)

```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.columns = [col.strip() for col in df.columns]
```

```python
permanent_province_counts = df['Permanent Province'].value_counts()
plt.figure(figsize=(10, 6))
permanent_province_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Permanent Province')
plt.ylabel('')
plt.show()
```

```python
reference_counts = df['refrence'].value_counts()
plt.figure(figsize=(10, 6))
reference_counts.plot.bar()
plt.title('Count of References')
plt.xlabel('Reference')
plt.ylabel('Count')
plt.show()
```

```python
temporary_province_counts = df['Temporary Province'].value_counts()
plt.figure(figsize=(10, 6))
permanent_province_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Distribution of  Province')
plt.ylabel('')
plt.show()
```
